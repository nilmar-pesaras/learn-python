```python
# Basic Problem

# Problem 1
# 1. Write a program that reverses a string.
# 2. Create a function that counts vowels in a string.
# 3. Write a program that checks if a string is a palindrome.

def reverse_string(string_input) -> str:
    return string_input[::-1]


def count_vowels(string_input,) -> int: 
    vowels = "aeiouAEIOU"
    return sum(1 for char in string_input if char in vowels)


def is_palindrome(text: str) -> str | None:
    cleaned = "".join(char.lower() for char in text if char.isalnum())
    if cleaned == cleaned[::-1]:
        return text  # Return the original input
    return None  # Return None if not a palindrome


input_string: str = "The quick brown fox jumps over the lazy dog"

print(f"reverse_string: {reverse_string(input_string)}")
print(f"count_vowels: {count_vowels(input_string)}")
print(f"is_palindrome: {is_palindrome(input_string)}")

```

```python
# Problem 2

# Create a function that alternates the case of each character in a string.


def alternate_case(string_input) -> str:
    result: str = ""
    for i, char in enumerate(string_input):
        if i % 2 == 0:
            result += char.upper()
        else:
            result += char.lower()
    return result


input_string: str = "The quick brown fox jumps over the lazy dog"

print(f"alternate_case: {alternate_case(input_string)}")
```

```python
# Clean Code Solution of Intermediate Problems No. 1 

def remove_special_characters(text: str, special_chars: str = "@#$%^&*+") -> str:
    """Remove special characters from the text."""
    translation_table = str.maketrans("", "", special_chars)
    return text.translate(translation_table)

def normalize_spaces(text: str) -> str:
    """Remove extra spaces and normalize whitespace."""
    return " ".join(text.split())

def capitalize_sentences(text: str) -> str:
    """Capitalize the first letter of each sentence."""
    sentences = [sentence[0].upper() + sentence[1:] for sentence in text.split(". ")]
    return ". ".join(sentences)

def clean_text(text: str) -> str:
    """Clean the text by removing special characters, normalizing spaces, and capitalizing sentences."""
    text = remove_special_characters(text)
    text = normalize_spaces(text)
    text = capitalize_sentences(text)
    return text

# Input string
input_string: str = ''' 
                    The quick brown fox jumps over the lazy dog. Python is a powerful language for data analysis and machine learning. It has a simple syntax and a large community of developers.
                    Hello World! This is a simple example of gibberish text. You need to remove all the special characters and clean it up. The final result should be a clean and readable text.
                    '''

# Clean the text
cleaned_text = clean_text(input_string)
print(cleaned_text)
```

```python
# Intermediate Problems No. 2

# Replace vowels with numbers using translate

# Translate Guide
# str.maketrans(x, y, z)
# x: string to be replaced
# y: string to replace with
# z: string to be removed


def replace_vowels_with_numbers(string_input) -> str:
    vowels = "aeiouAEIOU"
    return string_input.translate(str.maketrans(vowels, "1234512345"))

input_string: str = "The quick brown fox jumps over the lazy dog"
print(f"replace_vowels_with_numbers: {replace_vowels_with_numbers(input_string)}")
```

```python
# Intermediate Problems No. 3
# Remove Digits

def remove_digits(string_input) -> str:
    # Create a translation table
    translation_table = str.maketrans("", "", "0123456789")
    # Apply the translation
    return string_input.translate(translation_table)

input_string: str = "The quick3 brown fox jumps over the lazy dog"
print(f"remove_digits: {remove_digits(input_string)}")

```

```python
# Intermediate Problems No. 4
# Count word frequency

from collections import Counter
import re

def word_frequency(sentence: str) -> dict:
    words = re.findall(r"\b\w+\b", sentence.lower())
    return dict(Counter(words))

input_string: str = "The quick brown fox jumps over the lazy dog. Python is a powerful language for data analysis and machine learning. It has a simple syntax and a large community of developers."
print(f"word_frequency: {word_frequency(input_string)}")
```

```python
# Another solution for Intermediate Problems No. 4

from collections import Counter
import string

def word_frequency(sentence: str) -> dict:
    # Remove punctuation and convert to lowercase
    translator = str.maketrans("", "", string.punctuation)
    cleaned_sentence = sentence.translate(translator).lower()
    
    # Split into words and count frequency
    words = cleaned_sentence.split()
    return dict(Counter(words))

# Test the function
input_string: str = "The quick brown fox jumps over the lazy dog. Python is a powerful language for data analysis and machine learning. It has a simple syntax and a large community of developers."
print(f"word_frequency: {word_frequency(input_string)}")
```

```python
# Intermediate Problems No. 6
# Remove duplicate words

def remove_duplicate_words(sentence: str) -> str:
    words = sentence.split()
    return " ".join(set(words))

input_string: str = "The quick brown fox jumps over the lazy dog. The quick brown fox jumps over the lazy dog."
print(f"remove_duplicate_words: {remove_duplicate_words(input_string)}")
```

```python
# Intermediate Problems No. 7
# Count Substring Occurences

def count_substring_occurrences(text: str, substring: str) -> int:
    return text.count(substring)

input_string: str = "The quick brown fox jumps over the lazy dog. The quick brown fox jumps over the lazy dog."
print(f"count_substring_occurrences: {count_substring_occurrences(input_string, 'dog')}")
```

```python
# Intermediate Problems No. 8
# Format Data String

def format_date(date: str) -> str:
    year, month, day = date.split("-")
    return f"{day}/{month}/{year}"

input_string: str = "2024-01-01"
print(f"format_date: {format_date(input_string)}")
```

```python
# Advance Problems No. 1

def remove_special_characters(text: str, special_chars: str = "@#$%^&*+") -> str:
    """Remove all special characters from the text."""
    for char in special_chars:
        text = text.replace(char, "")
    return text

def normalize_spaces(text: str) -> str:
    """Normalize spaces by removing extra whitespace."""
    return " ".join(text.split())

def capitalize_sentences(text: str) -> str:
    """Capitalize the first letter of each sentence."""
    sentences = [sentence[0].upper() + sentence[1:] for sentence in text.split(". ")]
    return ". ".join(sentences)

def clean_paragraph(paragraph: str) -> str:
    """Clean a single paragraph."""
    paragraph = remove_special_characters(paragraph)
    paragraph = normalize_spaces(paragraph)
    paragraph = capitalize_sentences(paragraph)
    return paragraph

def clean_text(text: str) -> str:
    """Clean the entire text while preserving paragraphs."""
    paragraphs = text.split("\n\n")
    cleaned_paragraphs = [clean_paragraph(paragraph) for paragraph in paragraphs]
    return "\n\n".join(cleaned_paragraphs)

# Input text
gibberish_text = """
Th@e qu!ck br#own f$ox j%umps o^ver t&he l*azy d+og. Pyth@on is a p#owerful l$anguage f%or d^ata a&nalysis a*nd m+achine l@earning. I#t h$as a s%imple s^yntax a&nd a l*arge c+ommunity o@f d#evelopers.

H@ello W#orld! T$his is a s%imple e^xample o&f g*ibberish t+ext. Y@ou n#eed t$o r%emove a^ll t&he s*pecial c+haracters a@nd c#lean i$t u%p. T^he f&inal r*esult s+hould b@e a c#lean a$nd r%eadable t^ext.
"""

# Clean the text
cleaned_text = clean_text(gibberish_text)
print(cleaned_text)
```
