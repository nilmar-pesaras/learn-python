**Activity 2**

A project manager is overseeing an ICT project spanning 20 weeks. The team is tracking key performance indicators (KPIs) over time, such as Project Cost, Task Completion, Risk Level, Stakeholder Satisfaction, Budget Utilization, and Issue Count. Please provide lineplots for these KPIs. Use the excel file:  ICT_Project_Data_with_Dates.

```python
# Enable inline plotting
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
df = pd.read_csv("ICT_Project_Data_with_Dates.csv")

# Print information about the DataFrame
print("DataFrame columns:", df.columns.tolist())
print("\nFirst few rows of data:")
print(df.head())

# Convert Date column to datetime
df["Date"] = pd.to_datetime(
    df["Date"], dayfirst=True
)  # Use dayfirst=True if dates are in DD/MM/YYYY format

# Create a figure with subplots (3 rows x 2 columns)
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
fig.suptitle("ICT Project KPIs Over Time", fontsize=16, y=1.02)

# Get the actual column names from your DataFrame
kpis = [
    'Project_Cost',          
    'Task_Completion',       
    'Risk_Level',           
    'Stakeholder_Satisfaction',
    'Budget_Utilization',
    'Issue_Count'
]

# Colors for the lines
colors = ["blue", "green", "red", "purple", "orange", "brown"]

# Plot each KPI
for idx, (ax, kpi, color) in enumerate(zip(axes.flat, kpis, colors)):
    # Create line plot
    ax.plot(
        df["Date"],
        df[kpi],
        color=color,
        marker="o",
        linewidth=2,
        markersize=6,
        label=kpi,
    )

    # Customize each subplot
    ax.set_title(kpi, pad=10, fontsize=12)
    ax.set_xlabel("Date", fontsize=10)
    ax.set_ylabel("Value", fontsize=10)
    ax.grid(True, linestyle="--", alpha=0.7)

    # Rotate x-axis labels for better readability
    ax.tick_params(axis="x", rotation=45)

    # Add legend
    ax.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

**Activity 3**

Scenario: Analyzing Business Performance Metrics Using a Bar Plot
Context:
A company is evaluating the performance of 20 different business units (or projects) based on key financial, operational, and market indicators. The dataset includes Revenue, Expenses, Profit, Customer Satisfaction, Employee Count, Projects Completed, Defects Found, Support Tickets, Market Share, and Growth Rate.

Objective:
The company wants to use a bar plot to:

Compare Revenue and Expenses  Identify which business units are profitable or running at a loss.
Analyze Profit Margins  Understand which units generate the most profit.
Assess Market Share & Growth Rate  See which units are leading in the industry.
Evaluate Customer Satisfaction  Identify areas where customer experience needs improvement.
Track Employee Count vs. Projects Completed  Ensure teams are optimally staffed.

Use the Bar_Plot_DataFrame excel file.

```python
# Enable inline plotting
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
df = pd.read_csv("Bar_Plot_DataFrame.csv")

# Create subplots for each analysis
fig, axes = plt.subplots(3, 2, figsize=(20, 15))
fig.suptitle('Business Units Performance Analysis', fontsize=16, y=1.02)
            
               
# 1. Revenue vs Expenses
ax1 = axes[0, 0]
x = range(len(df['Category']))
width = 0.35
ax1.bar(x, df['Revenue'], width, label='Revenue', color='green', alpha=0.7)
ax1.bar([i + width for i in x], df['Expenses'], width, label='Expenses', color='red', alpha=0.7)
ax1.set_title('Revenue vs Expenses by Business Unit')
ax1.set_xlabel('Business Unit')
ax1.set_ylabel('Amount')
ax1.set_xticks([i + width/2 for i in x])
ax1.set_xticklabels(df['Category'], rotation=45)
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

# 2. Profit Margins
ax2 = axes[0, 1]
profit_bars = ax2.bar(df['Category'], df['Profit'], color='blue', alpha=0.7)
ax2.set_title('Profit by Business Unit')
ax2.set_xlabel('Business Unit')
ax2.set_ylabel('Profit')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True, linestyle='--', alpha=0.7)

# 3. Market Share and Growth Rate
ax3 = axes[1, 0]
ax3.bar(x, df['Market_Share'], width, label='Market Share', color='purple', alpha=0.7)
ax3.bar([i + width for i in x], df['Growth_Rate'], width, label='Growth Rate', color='orange', alpha=0.7)
ax3.set_title('Market Share vs Growth Rate')
ax3.set_xlabel('Business Unit')
ax3.set_ylabel('Percentage')
ax3.set_xticks([i + width/2 for i in x])
ax3.set_xticklabels(df['Category'], rotation=45)
ax3.legend()
ax3.grid(True, linestyle='--', alpha=0.7)

# 4. Customer Satisfaction
ax4 = axes[1, 1]
satisfaction_bars = ax4.bar(df['Category'], df['Customer_Satisfaction'], color='green', alpha=0.7)
ax4.set_title('Customer Satisfaction Ratings')
ax4.set_xlabel('Business Unit')
ax4.set_ylabel('Satisfaction Score')
ax4.tick_params(axis='x', rotation=45)
ax4.grid(True, linestyle='--', alpha=0.7)

# 5. Employee Count vs Projects Completed
ax5 = axes[2, 0]
ax5.bar(x, df['Employee_Count'], width, label='Employee Count', color='blue', alpha=0.7)
ax5.bar([i + width for i in x], df['Projects_Completed'], width, label='Projects Completed', color='red', alpha=0.7)
ax5.set_title('Employee Count vs Projects Completed')
ax5.set_xlabel('Business Unit')
ax5.set_ylabel('Count')
ax5.set_xticks([i + width/2 for i in x])
ax5.set_xticklabels(df['Category'], rotation=45)
ax5.legend()
ax5.grid(True, linestyle='--', alpha=0.7)

# Hide the last unused subplot
axes[2, 1].remove()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

**Activity 4**

Scenario: Employee Performance and Compensation Analysis Using Histograms & Scatter Plots
Context:
A company is analyzing its workforce data to understand how various factors, such as Age, Salary, Experience, Job Satisfaction, Performance Scores, and Annual Bonus, impact employee productivity and retention.

The dataset includes:

Demographics: Age, Experience
Compensation & Work Hours: Salary, Working Hours, Annual Bonus
Performance & Growth: Performance Score, Training Hours, Projects Handled, Certifications
Employee Satisfaction: Job Satisfaction
Objective:
The HR department wants to use histograms and scatter plots to:

Understand the Distribution of Employee Age, Salary, Experience, and Performance Scores  Histograms will help identify workforce demographics and salary trends.
Analyze the Relationship Between Salary & Experience  A scatter plot can show if higher experience leads to better salaries.
Investigate if More Training Hours Lead to Higher Performance  A scatter plot of Training Hours vs. Performance Score will reveal any correlation.
Assess Employee Satisfaction Based on Salary and Work Hours  Scatter plots of Job Satisfaction vs. Salary and Job Satisfaction vs. Working Hours will highlight factors affecting morale.

Use the excel file: Histogram___Scatter_Plot_DataFrame

```python
# Enable inline plotting
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# Import dataset
df = pd.read_csv("Histogram___Scatter_Plot_DataFrame.csv")

# Create figure with subplots
fig = plt.figure(figsize=(20, 15))

# Add main title at the top with proper spacing and formatting
plt.suptitle('Employee Performance and Compensation Analysis', 
             fontsize=20, 
             y=0.95,
             fontweight='bold')

# 1. Histograms for distributions
# Age distribution
plt.subplot(3, 3, 1)
plt.hist(df['Age'], bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.title('Age Distribution', pad=20)  # Added padding to avoid overlap
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)

# Salary distribution
plt.subplot(3, 3, 2)
plt.hist(df['Salary'], bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Salary Distribution', pad=20)  # Added padding to avoid overlap
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x/1000)}K'))

# Rest of the code remains the same...
# Experience distribution
plt.subplot(3, 3, 3)
plt.hist(df['Experience_Years'], bins=15, color='red', alpha=0.7, edgecolor='black')
plt.title('Experience Distribution', pad=20)  # Added padding to avoid overlap
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)

# Performance Score distribution
plt.subplot(3, 3, 4)
plt.hist(df['Performance_Score'], bins=15, color='purple', alpha=0.7, edgecolor='black')
plt.title('Performance Score Distribution', pad=20)  # Added padding to avoid overlap
plt.xlabel('Performance Score')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)

# 2. Scatter plot: Salary vs Experience
plt.subplot(3, 3, 5)
plt.scatter(df['Experience_Years'], df['Salary'], alpha=0.5, color='blue')
plt.title('Salary vs Experience', pad=20)  # Added padding to avoid overlap
plt.xlabel('Years of Experience')
plt.ylabel('Salary ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x/1000)}K'))

# 3. Scatter plot: Training Hours vs Performance Score
plt.subplot(3, 3, 6)
plt.scatter(df['Training_Hours'], df['Performance_Score'], alpha=0.5, color='green')
plt.title('Training Hours vs Performance', pad=20)  # Added padding to avoid overlap
plt.xlabel('Training Hours')
plt.ylabel('Performance Score')
plt.grid(True, linestyle='--', alpha=0.7)

# 4. Scatter plot: Job Satisfaction vs Salary
plt.subplot(3, 3, 7)
plt.scatter(df['Salary'], df['Job_Satisfaction'], alpha=0.5, color='red')
plt.title('Job Satisfaction vs Salary', pad=20)  # Added padding to avoid overlap
plt.xlabel('Salary ($)')
plt.ylabel('Job Satisfaction')
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x/1000)}K'))

# 5. Scatter plot: Job Satisfaction vs Working Hours
plt.subplot(3, 3, 8)
plt.scatter(df['Working_Hours'], df['Job_Satisfaction'], alpha=0.5, color='purple')
plt.title('Job Satisfaction vs Working Hours', pad=20)  # Added padding to avoid overlap
plt.xlabel('Working Hours')
plt.ylabel('Job Satisfaction')
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout with more space at the top for the main title
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Show plot
plt.show()
```
